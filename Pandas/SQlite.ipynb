```python
# 1. 테이블 생성 
cur = conn.cursor()
cur.execute('CREATE TABLE IF NOT EXISTS stats \
    (Back_no INT NOT NULL, \
     Name TEXT, \
     W INT, \
     L INT, \
     ERA float, \
     IP float, \
     SO INT, \
     PRIMARY KEY(Back_no));')
```

```python
# 2. 파이썬에서 'stats' csv 파일 읽기
stats = pd.read_csv('./pitchers.csv', encoding='EUC-KR')
stats
```

```python
# 3. 데이터베이스에 'stats' 테이블 추가
cur = conn.cursor()
sql = 'INSERT INTO stats VALUES (?, ?, ?, ?, ?, ?, ?);'
for i in range(12):
    cur.execute(sql, (
                      pitchers.iloc[i,0],
                      pitchers.iloc[i,1],
                      (int(pitchers.iloc[i,2])),
                      (int(pitchers.iloc[i,3])),
                      (float(pitchers.iloc[i,4])),
                      (float(pitchers.iloc[i,5])),
                      (int(pitchers.iloc[i,6]))))
conn.commit() # 변경사항 저장 
```

```python
# 데이터 테이블 확인
cur = conn.cursor()
cur.execute('SELECT * FROM stats')

for row in cur:
    print(row)
```

```python
# 4.2.1 데이터베이스에 있는 데이터 확인하기
sql = "SELECT p.Back_no, p.Name, p.Position, \
        s.ERA, s.W, s.L, s.IP, s.SO \
        FROM Philies0 AS p JOIN stats AS s \
        ON p.Name like s.Name;"

cur = conn.cursor()
cur.execute(sql)
rowp = []
for row in cur:
    rowp.append(row)
rowp
```

```python
# 연습문제 1
columnp = ["등번호","이름","포지션","ERA","승","패","이닝","삼진"]

           
p_data = pd.DataFrame(rowp,columns= columnp)
p_data
```

#### 연습문제 2
다음의 지시대로 DB 테이블을 만들고 이를 조회하는 프로그램을 만드시오.
국내의 대표적인 걸그룹 또는 보이그룹 5개 이상에 대하여 다음과 같은 정보를 갖는 테이블을 만드시오. *는 
- Primary Key id(*), group_name, 구성원 수, 데뷔일자, 소속사

이들이 불렀던 노래 또는 다른 사람이 불렀던 노래 10곡 이상에 대하여 다음의 정보를 갖는 테이블을 만드시오.

- song_id(*), song_name, 그룹 id, 발표년도, 작곡가, 도입부 가사

위 두개의 테이블을 조인한 결과를 가지고 다음의 필드를 갖는 데이터 프레임을 만드시오.그룹 이름, 구성원 수, 데뷔 일자, 노래 이름, 발표 년도


```python
groups = pd.read_csv('groups.csv', encoding='euc-kr')
groups 
songs = pd.read_csv('songs.csv',encoding='euc-kr')
songs
```

```python
cur.execute('DROP TABLE groups')
```

```python
cur = conn.cursor()
cur.execute('CREATE TABLE groups \
              (id INT NOT NULL, \
              Name TEXT, \
              Party INT NOT NULL, \
              debut datetime, \
              company TEXT, \
              PRIMARY KEY(id));')
```

```python
cur = conn.cursor()
cur.execute('CREATE TABLE songs \
              (song_id INT NOT NULL, \
              song_name TEXT, \
              group_id TEXT, \
              발표년도 datetime, \
              작곡가 TEXT, \
              도입부 가사 TEXT, \
              PRIMARY KEY(song_id));')
```

```python
cur = conn.cursor()
sql = 'INSERT INTO groups VALUES (?, ?, ?, ?, ?);'
for i in range(5):
    cur.execute(sql, (int(groups.iloc[i,0]),
                     groups.iloc[i,1],int(groups.iloc[i,2]),
                     groups.iloc[i,3],groups.iloc[i,4]))
conn.commit()
```

```python
cur = conn.cursor()
sql = 'INSERT INTO songs VALUES (?, ?, ?, ?, ?, ?);'
for i in range(10):
    cur.execute(sql, (int(songs.iloc[i,0]),
                     songs.iloc[i,1],songs.iloc[i,2],
                     songs.iloc[i,3],songs.iloc[i,4],
                     songs.iloc[i,5]))
conn.commit()
```

```python
cur = conn.cursor()
cur.execute('SELECT * FROM groups')

for row in cur:
    print(row)
```

```python
sql = "SELECT g.name, g.Party, \
        g.debut, g.company, s.song_name, s.발표년도 \
        FROM groups AS g JOIN songs AS s \
        ON g.name like s.group_id;"

cur = conn.cursor()
cur.execute(sql)
rowg = []
for row in cur:
    rowg.append(row)
rowg
```

```python
columns = ["그룹이름","구성원 수","데뷔 일자","회사명","노래 이름","발표 년도"]

           
g_data = pd.DataFrame(rowg,columns= columns)
g_data
```

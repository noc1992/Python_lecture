```python
from keras.models import Sequential
from keras.layers.core import Dense
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import train_test_split

import seaborn as sns
import numpy as np
import pandas as pd
import tensorflow as tf
```

```python
from keras.callbacks import ModelCheckpoint, EarlyStopping
import matplotlib.pyplot as plt
```

```python
data_pre = pd.read_csv('./dataset/wine.csv', header=None)
data = data_pre.sample(frac=1)
```

```python
print(data.head(5))
```

```python
seed = 0
np.random.seed(seed)
tf.set_random_seed(seed)
```

```python
dataset = data.values
X = dataset[:,0:11]
Y = dataset[:,11]
```

```python
sess = tf.Session()
Y_Hot = tf.one_hot(Y,depth=8).eval(session = sess)
Y_Hot
```

## 32 x 24 x 16 x 8 + sigmoid

```python
model1 = Sequential()
model1.add(Dense(32, input_dim=11, activation='relu'))
model1.add(Dense(24, activation='relu'))
model1.add(Dense(16, activation='relu'))
model1.add(Dense(8, activation='sigmoid'))
```

```python
model1.compile(loss='categorical_crossentropy',
             optimizer='adam',
             metrics=['accuracy'])
```

```python
model1.fit(X, Y_Hot, epochs=500, batch_size=500)
```

```python
df1 = data_pre.sample(frac=0.15)
```

```python
history1 = model1.fit(X, Y_Hot, validation_split=0.33, epochs=1500,
                   batch_size=500)
```

```python
import matplotlib.pyplot as plt

y_vloss1=history1.history['val_loss']
y_acc1=history1.history['acc']
```

```python
x_len1 = np.arange(len(y_acc1))
plt.plot(x_len1, y_vloss1, "o", c="red", markersize=3)
plt.plot(x_len1, y_acc1, "o", c="blue", markersize=3)
```
